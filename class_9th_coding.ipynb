```python
# coordinate tracking program
x1,y1 = 0,0 #starting position (origin)
x2,y2 = 5,6 #ending position

#calculating displacement using math logic
horizontal_distance= x2-x1
vertical_distance= y2-y1

#the pythagorean formula = distance squareel= a^2+b^2
distance_squareel = (horizontal_distance**2)+(vertical_distance**2)
total_distance = distance_squareel**0.5 #square root

print(f'the object moved from ({x1},{y1}) to ({x2},{y2}).')
print(f' the short-line distance (displacement) is:{total_distance} units:')
```
